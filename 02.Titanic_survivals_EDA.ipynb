```python
import pandas as pd

Data = pd.read_csv(
    "https://raw.githubusercontent.com/evgpat/edu_stepik_practical_ml/main/datasets/titanik.csv",
    sep=";"
    )
```

```python
Data.describe() # Посмотреть статистику по численным переменным
```

```python
Data.describe(include = 'object') # Посмотреть статистику по категориальным переменным
```

```python
Data['Sex'].unique() # Посмотреть, какие значения принимает признак Sex
```

```python
Data = Data[Data['Sex'] != 'unknown'] # Удаляем из таблички объекты с неопределенным полом
```

```python
Data['Sex'] = Data['Sex'].map({'male' : 1, 'female' : 0}) # Переводим категориальный признак Sex в числовой
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.barplot(x='Sex', y='Survived', data=Data, hue='Sex', palette='summer', legend=False) # Делаем bar plot диаграмму пол - выживаемость
plt.title('Sex - Survived')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', hue = 'Pclass', data=Data, palette='autumn') # Делаем bar plot диаграмму пол, класс билета - выживаемость
plt.title('Sex - Survived')
plt.show();
```

```python
corr = Data.corr(numeric_only=True) # Вычисление корреляционной матрицы
```
